```python
import numpy as np
```

```python
x = np.linspace(-10,10,100) # 두개의 패널 중 첫번째 패널을 생성하고 현재 축axis를 설정
```

```python
plt.subplot (2,1,1)
plt.plot(x, np.sin(x)) # 2행 1열 구조 그리드의 첫번째 패널에 사인 그래프 작성
```

```python
plt.subplot(2,1,2)
plt.plot(x, np.cos(x)); # 2행 1열 구조 그리드의 두번째 패널에 코사인 그래프 작성
```

**객체 지향 스타일 인터페이스**
- 플롯 그리드를 생성
- Ax는 두개의 축 객체의 배열 형성 -> *fig, ax=plt.subplots(2,1)*
 - (nrows, ncols) 2행 1열의 축 객체 생성
- 적절한 객체의 plot() 메소드를 호출
 - *ax[0].plot(x, np.sin(x))*
 - *ax[1].plot(x, np.cos(x));*

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

```python
# **2) 플롯 조정 - 선 색과 스타일**
plt.plot(x, np.sin(x-0), color='blue') # 색상을 이름으로 지정
```

```python
plt.plot(x, np.sin(x-1), color='g') # 짧은 색상코드 (rgb, cmyk)
```

```python
plt.plot(x, np.sin(x-2), color='0.75') # 회색조 지정, 0과 1사이 값
```

```python
plt.plot(x, np.sin(x-3), color='#FFDD44') # 16진수 코드 (RRGGBB, 00~FF 사이)
```

```python
plt.plot(x, np.sin(x-4), color=(1.0, 0.2, 0.3)) # RGB튜플, 0과 1사이 값
```

```python
plt.plot(x, np.sin(x-5), color='chartreuse'); # 모든 HTML 색상 이름을 지원
```

```python
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted'); # 키워드 사용
```

```python
plt.plot(x, x+4, linestyle='-') # solid와 동일한 단축 기호
plt.plot(x, x+5, linestyle='--') # dashed와 동일한 단축 기호
plt.plot(x, x+6, linestyle='-.') # dashdot과 동일한 단축 기호
plt.plot(x, x+7, linestyle=':'); # dotted와 동일한 단축 기호
```

```python
plt.plot(x, x+0, '-g') # solid green
plt.plot(x, x+1, '--c') # dashed cyan
plt.plot(x, x+2, '-.k') # dashdot black
plt.plot(x, x+3, ':r'); # dotted red
```

```python
# 축의 범위 지정
plt.plot(x, np.cos(x))
plt.xlim(-1,11)
plt.ylim(-1.5,1.5);
```

```python
#축의 상하좌우 반전시키기 -> 인수의 순서 바꾸면 된다
plt.plot(x, np.cos(x))
plt.xlim(11,-1)
plt.ylim(1.5,-1.5);
```

```python
# 축의 범위 지정
plt.plot(x, np.cos(x*2))
plt.axis([-1,11,-1.5,1.5]); # 인자로 [x축 최소값, x축 최대값, y축 최소값, y축 최대값] 쌍을 리스트로 전달
```

```python
# 축의 스케일 변경
plt.plot(x, np.exp(-x));
```

```python
plt.plot(x, np.exp(-x)) 
plt.yscale("log"); # y축을 "log"변환한것 (log, linear, symlog, logit 등이 들어갈 수 있다)
```

```python
# 축의 레이블 추가
plt.plot(x, np.cos(x))
plt.title("A cosine Curve") # 그림의 제목 지정
plt.xlabel("x")
plt.ylabel("sin(x)"); # 축의 레이블 지정
```
